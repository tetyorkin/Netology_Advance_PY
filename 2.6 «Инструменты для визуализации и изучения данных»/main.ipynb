```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 12,9

# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Читаем датасет и смотрим, что он из себя представляет.

```python
girls = pd.read_csv('out.csv')
girls
```

В данном датасете 214125 записей разбитых на 9 колонок. Посторим гистограмму распределения по возрасту

```python
girls.hist(column=['Age'], bins=33)
plt.title('Гистограмма соотношение количества к возрасту')
plt.xlabel('Возраст')
plt.ylabel('Количество')
```


Из данной гистограммы можно сделать вывод, что самое большое количество девушек эскорт услуг в г. Москва в возрасте 25 и 26 лет, около 24 тысяч анкет. Далее при увеличении возраста, количество анкет падает, но далее каждые 5 лет наблюдается рост количества предложений от среднего числа. Также отсутвуют анкеты представительниц низкой социальной отвественности старше 50 лет.

Построим boxplot распределения по росту девушек.

```python
height_girls = girls['Height']
height_girls.plot(kind='box')
plt.ylabel('Рост в см.')
plt.grid()
```

Медиана распределения boxplot указывает, на средний рост девушек экскорт услуг 170 см. достаточно высокие девушки. Усы boxplot показывают минимальное и максимальное значение выборки. В нашей выборке минимальное значение 158 см. и максимальное 181 см. В выборке присутствует большое количество выбросов выше и ниже усов, что говорит о присутвиии в выборке низких (ниже 158 см.) и высоких (выше 181) девушках, но в маленьких количествах. 

Построим круговую диаграмму сравнения количества топ-10 цен по всем предложениям

```python

girls['av_price'] = girls['Price_USD']
girls['av_price'].value_counts().head(10).plot.pie()
```

Из круговой диаграммы видно, что самым большим спросом полюзуются девушки предложением следующих цен: 45$, 37$ и 30$

Теперь посмотрим корреляцию, есть ли связь между возрастом и ценой

```python
sns.jointplot(x='Price_USD', y='Age', 
              data=girls, kind='scatter')
```

По графику можно сделать вывод что в среднем цена не сильно завистит от возраста.

Та же самая визуализация с добавлением линии регресии

```python
sns.jointplot(x='Price_USD', y='Age', 
              data=girls, kind='reg')
```
